```python
import numpy as np
import pandas as pd
import ast
```

```python
movies = pd.read_csv("C:\\Users\\alluh\\Downloads\\tmdb_5000_movies.csv\\tmdb_5000_movies.csv")
credit = pd.read_csv("C:\\Users\\alluh\\Downloads\\tmdb_5000_credits.csv\\tmdb_5000_credits.csv")
```

```python
merged=movies.merge(credit, left_on='id', right_on='movie_id')
```

```python
merged.info()
```

```python
merged = merged[['id', 'title_x', 'keywords', 'genres', 'overview', 'cast', 'crew']]
merged.head()
```

```python
merged.isnull().sum()
```

```python
merged.dropna(inplace=True)
```

```python
merged.duplicated().sum()
```

```python
def convert(obj):
    L = []
    for x in ast.literal_eval(obj):
        L.append(x['name'])
    return L
```

```python
merged['genres']=merged['genres'].apply(convert)
```

```python
merged['keywords']=merged['keywords'].apply(convert)
```

```python
merged.head()
```

```python
def convert2(obj):
    L = []
    count = 0
    for x in ast.literal_eval(obj):
        if count != 3:
            L.append(x['name'])
            count = count + 1
        else:
            break
    return L
```

```python
merged['cast']=merged['cast'].apply(convert2)
```

```python
merged.head()
```

```python
def convert3(obj):
    L = []
    for x in ast.literal_eval(obj):
        if x['job'] == 'Director':
            L.append(x['name'])
            break
    return L
```

```python
merged['crew']=merged['crew'].apply(convert3)
```

```python
merged.head()
```

```python
merged['overview']=merged['overview'].apply(lambda x:x.split())
```

```python
merged.head()
```

```python
def space(li):
    L = []
    for i in li:
        L.append(i.replace(" ", ""))
    return L
merged['genres']=merged['genres'].apply(space)
merged['keywords']=merged['keywords'].apply(space)
merged['cast']=merged['cast'].apply(space)
merged['crew']=merged['crew'].apply(space)
```

```python
merged.head()
```

```python
merged['tags'] = merged['overview'] + merged['keywords'] + merged['cast'] + merged['crew'] + merged['genres']
merged.head()
```

```python
df = merged[['id', 'title_x', 'tags']]
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x : " ".join(x))
df.head()
```

```python
df['tags'][0]
```

```python
df['tags'] = df['tags'].apply(lambda x : x.lower())

df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem (txt):
    y= []
    for i in txt.split():
        y.append(ps.stem(i))
    return " ".join(y)


df['tags'] = df['tags'].apply(stem)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vect = cv.fit_transform(df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similar = cosine_similarity(vect)
```

```python
cosine_similarity(vect).shape
```

```python
similar[0]
```

```python
def recommend(title):
    cleaned_title = title.replace(" ", "").lower()
    
    matching_rows = df[df['title_x'].str.replace(" ", "").str.lower() == cleaned_title]
    
    if not matching_rows.empty:
        index = matching_rows.index[0]
        movie_simi = similar[index]
        movies = sorted(list(enumerate(movie_simi)), reverse=True, key=lambda x: x[1])[1:6]

        for ids in movies:
            print(df.iloc[ids[0]]['title_x'])
    else:
        print('No movie in the Database')
```

```python
recommend('avatar')
```
